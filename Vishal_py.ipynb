```python
import numpy as np

#PANDA SERIES AND SERIES DATA MANIPULATION
#FROM LIST
s1= pd.Series([10,20,30,40])
#From list with custom index
s2 = pd.Series([100,200,300],index=['a','b','c'])
#From dictionary
s3 = pd.Series({'x':5,'y':10,'z':15})
print(s1)
print(s2)
print(s3)



#Slicing & indexing

print(s2['b'])
print(s2[1])
print(s2[['a','c']])
print(s2[0:2])

#MODIFYING SERIES

s2['b'] = 250  #Update value
s2['d'] = 400  #Add new item
s2.drop('a',inplace=True)
print(s2)

#HANDLING MISSING VALUES (NANS)
s= pd.Series([10,None,30,np.nan])

print(s.isnull())   #True where NaN

print(s.notnull())  #Inverse

s.fillna(0)  #Replace NaN with 0

s.dropna()   #Remove NaN entries

#DESCRIPTIVE STATISTICS

s = pd.Series([10,20,30,40,50,])

print(s.sum()) # 150
print(s.mean()) #30.0
print(s.min(),s.max()) #10 50

#STRING METHODS ON SERIES
s= pd.Series([30,10,40,20]),index=(['d','b','a','c'])
print(s.sort_values()) # Sort by value
print(s.sort_index())  #Sort by index
print(s.sort_index())  #Rank based


```
